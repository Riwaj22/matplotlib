```python
import dash
import dash_core_components as dcc
import dash_html_components as html
```

```python
app = dash.Dash()
```

```python
colors = {
    'text':'#ff0000',
    'plot_color':'#808080',
    'paper_color':'#808080'
}
```

```python
import numpy as np
import plotly.graph_objs as go
x = np.random.randint(1,10,60)
y = np.random.randint(10,20,60)
```

```python
app.layout = html.Div([
    
    
    dcc.Graph(
        id = 'scatter plot',
        figure={
            'data': [
                go.Scatter(
                    x = x,
                    y = y,
                    mode = 'markers'
                    
                )
            ],
            'layout': go.Layout(
                title = 'Scatter Plot of Random 60 Numbers',
                xaxis = {'title':'Random x values'},
                yaxis = {'title': 'Random Y values'}
                
            )
        }

        
        
        
    )
        
          
    ])
```

```python
if __name__ == '__main__':
    app.run_server(port = 4050)
```
